```python
soup = BeautifulSoup(site.text, 'html.parser')
```

```python
tables = soup.find_all('tbody')

competition_data = []
for table in tables:
    trs = table.find_all('tr')
    for tr in trs:
        ths = tr.find_all('th')
        for th in ths:
            competition_name = th.string
            link = th.a['href']
            competition_data.append({'Competition Name': competition_name, 'Link': f"https://fbref.com/{link}"})

all_comps = pd.DataFrame(competition_data)
all_comps.to_csv('all competitions.csv')
```
